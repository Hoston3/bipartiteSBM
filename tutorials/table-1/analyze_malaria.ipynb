```python
import pickle
import glob
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (6,6)
```

```python
files = glob.glob('./malaria/*.pickle')
```

```python
mdl = []
k = []

for f in files:
    data = pickle.load(open(f, 'rb'))
    mdl += [(data["mdl"] / data["e"], data["gt_mdl"] / data["e"], data["gt_h_mdl"] / data["e"])]
    k += [((data["ka"], data["kb"]), (data["gt_ka"], data["gt_kb"]), (data["gt_h_ka"], data["gt_h_kb"]))]
```

```python
for k_ in k:
    for idx, (x, y) in enumerate(k_):
        if idx == 0:
            kwargs = {"c": "#0074D9", "label": "YEN"}
        elif idx == 1:
            kwargs = {"c": "#39CCCC", "label": "flat GT"}
        else:
            kwargs = {"c": "#85144b", "label": "nested GT"}
        plt.scatter(x, y, **kwargs)
plt.xlabel("ka")
plt.ylabel("kb")
```

```python
for idx, _mdl in enumerate(mdl):
    m = _mdl[0]
    kwargs = {"c": "#0074D9", "label": "YEN"}
    plt.scatter(idx, m, **kwargs)
    m = _mdl[1]
    kwargs = {"c": "#39CCCC", "label": "flat GT"}
    plt.scatter(idx, m, **kwargs)
    m = _mdl[2]
    kwargs = {"c": "#85144b", "label": "nested GT"}
    plt.scatter(idx, m, **kwargs)
    
plt.ylabel("Description length (per edge)")
plt.xlabel("Instance of Trials (out of 19)")
```
